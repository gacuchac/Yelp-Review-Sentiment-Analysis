```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
```

```python
temp_df = pd.DataFrame(columns=['stars','text'])
with pd.read_json("yelp_academic_dataset_review.json", orient="records", lines=True, chunksize=10000) as reader:
    for chunk in reader:
        aux = chunk.drop(chunk[chunk['stars']==3].index)[['stars','text']]
        temp_df = pd.concat([temp_df, aux])

#reviews = pd.DataFrame([pd.Series(stars), pd.Series(text), pd.Series(label)])
#reviews.head()
    
```

```python
temp_df.to_csv("filtered_reviews.csv")
```

```python
temp_df["label"] = [1 if s > 3 else 0 for s in temp_df["stars"]]
```

```python
X_train, X_test, y_train, y_test = train_test_split(temp_df[['stars','text']], temp_df['label'], test_size=0.1, train_size=0.4)
```

```python
X_train.to_csv('X_train.csv')
y_train.to_csv('y_train.csv')
X_test.to_csv('X_test.csv')
y_test.to_csv('y_test.csv')
```
